# Notebook auxiliar

En este cuadernillo auxiliar se estudiarán la mejora en la correlación obtenida por el proceso de centrado de la temperatura en función de la normalización de los datos de demanda de energía eléctrica y el exponente del centrado.

```python
from modulo_datos import estudio_temperatura
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dtype={'Indice': np.int64,
       'Valor': np.float64,
       'Prevision': np.float64,
       'Dia': str,
       'Dia_Semana': str,
       'Hora': str,
       'Temperatura': np.float64,
       'Fecha': str,
       'Valor_Normalizado1': np.float64,
       'Valor_Normalizado2': np.float64,
       'Valor_Normalizado3_7d': np.float64,
       'Valor_Normalizado3_1d': np.float64,
       'Temperatura_Centrada0': np.float64,
       'Temperatura_Centrada1': np.float64,
       'Temperatura_Centrada2': np.float64,
       'Temperatura_Centrada3_7d': np.float64,
       'Temperatura_Centrada3_1d': np.float64}

data = pd.read_csv('data.csv', sep=';', index_col=0, dtype=dtype, parse_dates=['Fecha'])

l_t = [10.0+k for k in range(16)]
```

```python
l_corr0, cor0 = estudio_temperatura(data['Temperatura'], data['Valor'], l_t)
l_corr1, cor1 = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado1'], l_t)
l_corr2, cor2 = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado2'], l_t)
l_corr37, cor37 = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado3_7d'], l_t)
l_corr31, cor31 = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado3_1d'], l_t)

l2_corr0, _ = estudio_temperatura(data['Temperatura'], data['Valor'], l_t, pot=2)
l2_corr1, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado1'], l_t, pot=2)
l2_corr2, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado2'], l_t, pot=2)
l2_corr37, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado3_7d'], l_t, pot=2)
l2_corr31, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado3_1d'], l_t, pot=2)

l3_corr0, _ = estudio_temperatura(data['Temperatura'], data['Valor'], l_t, pot=3)
l3_corr1, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado1'], l_t, pot=3)
l3_corr2, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado2'], l_t, pot=3)
l3_corr37, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado3_7d'], l_t, pot=3)
l3_corr31, _ = estudio_temperatura(data['Temperatura'], data['Valor_Normalizado3_1d'], l_t, pot=3)
```

```python
ancho = 12
alto = 6

plt.figure(figsize=(ancho,alto))
plt.plot(l_t, l_corr0, '.', label="1")
plt.plot(l_t, [cor0 for _ in l_t])
plt.plot(l_t, l2_corr0, '.', label="2")
plt.plot(l_t, l3_corr0, '.', label="3")
plt.title('Correlacion Valor - Temperatura Centrada En T (Sin Normalizacion)')
plt.xlabel('T (ºC)')
plt.ylabel('correlacion')
plt.ylim([0.0, 0.4])
plt.legend()
plt.show()

plt.figure(figsize=(ancho,alto))
plt.plot(l_t, l_corr1, '.', label="1")
plt.plot(l_t, [cor1 for _ in l_t])
plt.plot(l_t, l2_corr1, '.', label="2")
plt.plot(l_t, l3_corr1, '.', label="3")
plt.title('Correlacion Valor - Temperatura Centrada En T (Con Normalizacion Por Aproximacion A Dia Laborable)')
plt.xlabel('T (ºC)')
plt.ylabel('correlacion')
plt.ylim([0.0, 0.4])
plt.legend()
plt.show()

plt.figure(figsize=(ancho,alto))
plt.plot(l_t, l_corr2, '.', label="1")
plt.plot(l_t, [cor2 for _ in l_t])
plt.plot(l_t, l2_corr2, '.', label="2")
plt.plot(l_t, l3_corr2, '.', label="3")
plt.title('Correlacion Valor - Temperatura Centrada En T (Con Normalizacion Por Semana Previa)')
plt.xlabel('T (ºC)')
plt.ylabel('correlacion')
plt.ylim([0.0, 0.4])
plt.legend()
plt.show()

plt.figure(figsize=(ancho,alto))
plt.plot(l_t, l_corr37, '.', label="1")
plt.plot(l_t, [cor37 for _ in l_t])
plt.plot(l_t, l2_corr37, '.', label="2")
plt.plot(l_t, l3_corr37, '.', label="3")
plt.title('Correlacion Valor - Temperatura Centrada En T (Con Normalizacion Mixta 7 Dias)')
plt.xlabel('T (ºC)')
plt.ylabel('correlacion')
plt.ylim([0.0, 0.4])
plt.legend()
plt.show()

plt.figure(figsize=(ancho,alto))
plt.plot(l_t, l_corr31, '.', label="1")
plt.plot(l_t, [cor31 for _ in l_t])
plt.plot(l_t, l2_corr31, '.', label="2")
plt.plot(l_t, l3_corr31, '.', label="3")
plt.title('Correlacion Valor - Temperatura Centrada En T (Con Normalizacion Mixta 1 Dia)')
plt.xlabel('T (ºC)')
plt.ylabel('correlacion')
plt.ylim([0.0, 0.4])
plt.legend()
plt.show()
```
